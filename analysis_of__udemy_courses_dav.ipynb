# **DETECTION AND HANDLING OF OUTLIERS**

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=data['num_subscribers'])
plt.xlabel('Number of Subscribers')
plt.ylabel('Value')
plt.title('Box Plot of Number of Subscribers')
plt.show()
```

```python
numerical_columns = ['price', 'num_subscribers', 'num_reviews', 'num_lectures', 'content_duration']

plt.figure(figsize=(10, 6))
data[numerical_columns].boxplot()
plt.title('Box Plot of Numerical Columns')
plt.xlabel('Numerical Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

df_cleaned = remove_outliers_iqr(data, numerical_columns)
data=df_cleaned
print("Summary statistics after removing outliers:")
print(data.describe())
```

```python
plt.figure(figsize=(10, 6))
data[numerical_columns].boxplot()
plt.title('Box Plot of Numerical Columns')
plt.xlabel('Numerical Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Correlation Matrix**

```python
# Drop non-numeric columns or encode them if possible
numeric_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix shows the relationship between different numerical features in the dataset. The values range from -1 to 1, where:

- 1 indicates a perfect positive correlation (as one variable increases, the other also increases).
- 0 indicates no correlation (changes in one variable do not affect the other).
- -1 indicates a perfect negative correlation (as one variable increases, the other decreases).

Here are some insights from the correlation matrix:

- There is a strong positive correlation between `price` and `num_subscribers`, which means that courses with higher prices tend to have more subscribers.
- There is a moderate positive correlation between `price` and `num_lectures`, which means that courses with more lectures tend to have higher prices.
- There is a moderate positive correlation between `num_subscribers` and `num_lectures`, which means that courses with more lectures tend to have more subscribers.
- There is a weak positive correlation between `content_duration` and `num_lectures`, which means that courses with more lectures tend to have longer content duration.
- There is a weak positive correlation between `content_duration` and `price`, which means that courses with higher prices tend to have longer content duration.

These insights can be used to understand the relationships between different features in the dataset and to make informed decisions about course pricing, content duration, and number of lectures. For example, if an instructor wants to increase the number of subscribers to their course, they could consider increasing the price or the number of lectures.

   - Beginner courses with longer durations seem to receive more reviews compared to shorter ones.
   - Intermediate and Advanced courses show a wider spread, with some longer courses receiving fewer reviews and vice versa.
   - All Levels courses display a similar pattern to Beginner courses, with longer courses generally receiving more reviews.
This analysis suggests that course creators aiming for higher engagement and feedback in the form of reviews may consider creating courses with longer durations
