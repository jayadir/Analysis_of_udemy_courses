# **NUMBER OF COURSES LAUNCHED PER YEAR**

```python
data.loc[:, 'Year'] = data['published_timestamp'].dt.year
```

```python
sns.countplot(x='Year', data=data)
```

# **CATEGORY WISE COUNT OF COURSES PER YEAR**

```python
data.groupby('Year')['subject'].value_counts()
```

```python
grouped_data = data.groupby(['Year', 'subject']).size().reset_index(name='count')

plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='count', hue='subject', data=grouped_data)
plt.title('Number of Courses by Subject Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Courses')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Subscription Trends Over Time**

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='published_timestamp', y='num_subscribers', data=data)
plt.title('Subscription Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45)
plt.show()
```

# **Price Distribution**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['price'], kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# **Number of Lectures vs. Course Duration**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='num_lectures', y='content_duration', data=data)
plt.title('Number of Lectures vs. Course Duration')
plt.xlabel('Number of Lectures')
plt.ylabel('Content Duration')
plt.show()
```

# **Correlation Matrix**

```python
# Drop non-numeric columns or encode them if possible
numeric_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

# **Content Duration vs Engagement**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='content_duration', y='num_reviews', hue='level', palette='Set1')
plt.title('Content Duration vs. Number of Reviews')
plt.xlabel('Content Duration (hours)')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()
```
